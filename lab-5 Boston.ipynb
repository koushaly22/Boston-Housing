```python
def load_boston():
    scaler = StandardScaler()
    boston = datasets.load_boston()
    X=boston.data
    y=boston.target
    X = scaler.fit_transform(X)
    return train_test_split(X,y)
```

```python
print train_test_split(X,y)
```

```python
X_train, X_test, y_train, y_test = load_boston()
```

<b>Fitting a Linear Regression

```python
clf = LinearRegression()
clf.fit(X_train, y_train)
```

```python
print('Coefficients: \n', clf.coef_)
```

```python
print('INtercept: \n', clf.intercept_)
```

<b>Making Prediction

```python
zip (y_test, clf.predict(X_test))
```

```python
mean_squared_error(y_test, clf.predict(X_test))
```

```python
r2_score(y_test, clf.predict(X_test))
```

<b>From this it is seen that
<li>MSE is 28.1638093614798
<li>r2_score is 0.7070490824283745

```python
#Now we will see Ridge linear regression 
#lets assign a value to alpha and instiantiate new object as aplpha=0.01
r = Ridge(alpha=.01)
r.fit(X_train, y_train)
```
